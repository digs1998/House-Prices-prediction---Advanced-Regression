# Advanced Housing Prices- Feature Engineering
**The main aim of this project is to predict the house price based on various features which we will discuss as we go ahead**

**Dataset to downloaded from the below link
https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data**

**We will be performing all the below steps in Feature Engineering**

1. **Missing values**
2. **Temporal variables**
3. **Categorical variables: remove rare labels**
4. **Standarise the values of the variables to the same range**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

**We will do the process of feature engineering on both train and test.csv files here**

```python
pd.options.display.max_columns

data = pd.read_csv('C:/Users/Lenovo/Downloads/house-prices-advanced-regression-techniques/train.csv')
data.head()
```

## Missing Values

```python
##First missing values in Categorical cols
features_with_na = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes=='O']
features_with_na
```

```python
##As we had found out, certain missing values have significant contribution to increased sales price, so we will be instead assigning labels to it
def replace_cat_feature(data, features_with_na):
    dataset = data.copy()
    dataset[features_with_na] = dataset[features_with_na].fillna('Missing')
    return dataset

data = replace_cat_feature(data ,features_with_na)

data[features_with_na].isnull().sum()
```

```python
data[features_with_na].head()
```

```python
##Now, for numerical features
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes!='O']
numerical_with_nan
```

```python
## The numerical columns, also other than having a skewed distribution, consisted of outliers.

for feature in numerical_with_nan:
    median_value = data[feature].median()
    #creating a new col for storing NaN values
    data[feature+'NaN'] = np.where(data[feature].isnull(),1,0)
    data[feature].fillna(median_value, inplace=True)
    
data[numerical_with_nan].isnull().sum()
```

## Temporal Variables

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    data[feature]=data['YrSold']-data[feature]
```

## Numerical Variables
- **Now, we shall apply log transformation to convert skewed data to normal distribution**

```python
data.head()
```

```python
num_vals = ['LotFrontage','LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_vals:
    data[feature] = np.log(data[feature])

data.head()
```

## Handling Rare Categorical Features
- Here, the categories, that constitute less than 1% of the total will be considered

```python
categorical_features=[feature for feature in data.columns if data[feature].dtype=='O']
```

```python
for feature in categorical_features:
    temp=data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df=temp[temp>0.01].index
    data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rare_var')
```

## Labeling these Categorical features

```python
for feature in categorical_features:
    labels_ordered = data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    data[feature] = data[feature].map(labels_ordered)
```

```python
## So now, this is our final dataset, we will now apply MinMax preprocessing to it
data.head()
```

## MinMax Scaler

```python
scaling_feature=[feature for feature in data.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)

mx = MinMaxScaler()
mx.fit(data[scaling_feature])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([data[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(mx.transform(data[scaling_feature]), columns=scaling_feature)],
                    axis=1)
```
